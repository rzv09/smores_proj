```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fl_file = '/Users/ramanzatsarenko/smores_proj/out/Fl_100epochs_10runs_2026-01-25_23-09-27/metrics.csv'
model_FL = 'model_FL_fl3fl4'
datasets = [['fl_dataset1'], ['fl_dataset2'], ['fl_dataset3'], ['fl_dataset4']]

def get_mare(datasets):
    mare = {}
    loss = {}
    base_MARE = {}

    df_tl = pd.read_csv(fl_file)

    for dataset in datasets:
        data = df_tl.loc[(df_tl['model'] == model_FL) & (df_tl['dataset'].isin(dataset))]
        this_mare = data['MARE']
        this_loss = data['SmoothL1Loss()']
        this_b_MARE = data['MARE_baseline']
        mare[dataset[0]] = this_mare
        loss[dataset[0]] = this_loss
        base_MARE[dataset[0]] = this_b_MARE


    
    return mare, loss, base_MARE

fl_mare, fl_loss, fl_b_mare = get_mare(datasets)
```

```python
fl_file = '/Users/ramanzatsarenko/smores_proj/out/Fl_100epochs_10runs_2026-01-07_12-09-37/metrics.csv'
tl_file = '/Users/ramanzatsarenko/smores_proj/out/tl_sbfl_100epochs_1runs_2025-11-10_12-10-58/metrics.csv'
model_TL = 'model_sb1sb2fl3fl4'
model_FL = 'model_FL_sb1sb2fl3fl4'
# dataset='fl_dataset1'
```

```python
def get_mare(datasets):
    tl_mares = {}
    fl_mares = {}

    df_fl = pd.read_csv(fl_file)
    df_tl = pd.read_csv(tl_file)

    for dataset in datasets:
        tl_data = df_tl.loc[(df_tl['model'] == model_TL) & (df_tl['dataset'].isin(dataset))]
        fl_data = df_fl.loc[(df_fl['model'] == model_FL) & (df_fl['dataset'].isin(dataset))]
        tl_mare = tl_data['MARE']
        fl_mare = fl_data['MARE']  
        tl_mares[dataset[0]] = tl_mare
        fl_mares[dataset[0]] = fl_mare
    
    return tl_mares, fl_mares
    
```

```python
tl_res, fl_res = get_mare([['fl_dataset1'], ['fl_dataset2'], ['sb_dataset3'], ['sb_dataset4']])

def extract_scalar(v):
    # handles pandas Series, numpy scalar, or float
    if hasattr(v, "values"):
        return float(v.values[0])
    return float(v)

fl_res = {
    'fl_dataset1': fl_res['fl_dataset1'],
    'fl_dataset2': fl_res['fl_dataset2'],
    'sb_dataset3': fl_res['sb_dataset3'],
    'sb_dataset4': fl_res['sb_dataset4'],
}

tl_res = {
    'fl_dataset1': tl_res['fl_dataset1'],
    'fl_dataset2': tl_res['fl_dataset2'],
    'sb_dataset3': tl_res['sb_dataset3'],
    'sb_dataset4': tl_res['sb_dataset4'],
}

datasets = list(fl_res.keys())

matrix = np.array([
    [extract_scalar(fl_res[d]) for d in datasets],
    [extract_scalar(tl_res[d]) for d in datasets],
])

methods = ["FL", "TL"]
```

```python
matrix
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 2.8))

im = ax.imshow(matrix, cmap="coolwarm")

# ticks
ax.set_xticks(np.arange(len(datasets)))
ax.set_yticks(np.arange(len(methods)))

ax.set_xticklabels(datasets, rotation=30, ha="right")
ax.set_yticklabels(methods)

# annotate cells
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        ax.text(j, i, f"{matrix[i, j]:.3%}",
                ha="center", va="center", fontsize=12)

ax.set_title("MARE comparison (lower is better)")
fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()
```

```python
percent_diff = (matrix[1] - matrix[0]) / matrix[0] * 100
fig, ax = plt.subplots(figsize=(8, 2.2))

im = ax.imshow(
    percent_diff.reshape(1, -1),
    cmap="coolwarm",
    vmin=-10,  # adjust if needed
    vmax=10
)

ax.set_xticks(range(len(datasets)))
ax.set_xticklabels(datasets, rotation=30, ha="right")
ax.set_yticks([0])
ax.set_yticklabels(["TL vs FL (%)"])

for j, v in enumerate(percent_diff):
    ax.text(j, 0, f"{v:+.2f}%", ha="center", va="center", fontsize=12)

cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Percentage difference (%)")

ax.set_title("Relative MARE difference (TL  FL) / FL")

plt.tight_layout()
plt.show()

```

```python
df["delta_mare"] = df["TL_MARE"] - df["FL_MARE"]
df["delta_loss"] = df["TL_LOSS"] - df["FL_LOSS"]
df["ratio_mare"] = df["TL_MARE"] / df["FL_MARE"]
df["ratio_loss"] = df["TL_LOSS"] / df["FL_LOSS"]

labels = df.index.tolist()
y = np.arange(len(labels))


# ============================================================
# Option 1: Dumbbell / connected-dot plots (FL vs TL)
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(12, 4), constrained_layout=True, sharey=True)

# MARE dumbbell
ax = axes[0]
for i, ds in enumerate(labels):
    ax.plot([df.loc[ds, "FL_MARE"], df.loc[ds, "TL_MARE"]], [i, i], linewidth=2)
ax.scatter(df["FL_MARE"], y, s=60, label="FL")
ax.scatter(df["TL_MARE"], y, s=60, label="TL")
ax.set_title("Dumbbell: FL vs TL (MARE)")
ax.set_xlabel("MARE")
ax.set_yticks(y)
ax.set_yticklabels(labels)
ax.invert_yaxis()  # top to bottom
ax.legend()

# LOSS dumbbell
ax = axes[1]
for i, ds in enumerate(labels):
    ax.plot([df.loc[ds, "FL_LOSS"], df.loc[ds, "TL_LOSS"]], [i, i], linewidth=2)
ax.scatter(df["FL_LOSS"], y, s=60, label="FL")
ax.scatter(df["TL_LOSS"], y, s=60, label="TL")
ax.set_title("Dumbbell: FL vs TL (LOSS)")
ax.set_xlabel("LOSS")
ax.legend()

plt.show()


# ============================================================
# Option 2: Ratio plots (TL/FL) with log y-scale
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(12, 4), constrained_layout=True)

x = np.arange(len(labels))

# Ratio MARE
ax = axes[0]
ax.plot(x, df["ratio_mare"].values, marker="o")
ax.axhline(1.0, linewidth=1)
ax.set_yscale("log")
ax.set_title("Ratio: TL/FL (MARE)")
ax.set_ylabel("TL / FL (log scale)")
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=30, ha="right")

# Ratio LOSS
ax = axes[1]
ax.plot(x, df["ratio_loss"].values, marker="o")
ax.axhline(1.0, linewidth=1)
ax.set_yscale("log")
ax.set_title("Ratio: TL/FL (LOSS)")
ax.set_ylabel("TL / FL (log scale)")
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=30, ha="right")

plt.show()


# ============================================================
# Option 3: Heatmap of deltas (TL - FL)
# ============================================================
heat = np.vstack([df["delta_mare"].values, df["delta_loss"].values]).T  # shape: (datasets, 2)

fig, ax = plt.subplots(figsize=(5.5, 4), constrained_layout=True)
im = ax.imshow(heat, aspect="auto")

ax.set_title("Heatmap:  = TL - FL")
ax.set_xticks([0, 1])
ax.set_xticklabels(["MARE", "LOSS"])
ax.set_yticks(np.arange(len(labels)))
ax.set_yticklabels(labels)

cbar = plt.colorbar(im, ax=ax)
cbar.set_label(" (TL - FL)")

plt.show()
```
