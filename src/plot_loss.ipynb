```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

SB_filepath = '/Users/ramanzatsarenko/smores_proj/out/crossval_SB_FL_100epochs_1runs_2025-10-29_12-53-19/metrics.csv'

def plot_ct(file, learning_type, ds):
    df = pd.read_csv(file)
    model1 = df.iloc[1]
    model2 = df.iloc[5]
    model3 = df.iloc[9]
    model4 = df.iloc[13]
    plt.figure(figsize=(10, 5))
    plt.plot(ast.literal_eval(model1['train_loss']), color='#1f77b4', label=f'{ds}1')
    plt.plot(ast.literal_eval(model2['train_loss']), color="#4cbb20", label=f'{ds}2')
    plt.plot(ast.literal_eval(model3['train_loss']), color="#f78318", label=f'{ds}3')
    plt.plot(ast.literal_eval(model4['train_loss']), color="#dc1515", label=f'{ds}4')
    plt.title(f'{learning_type} Loss')
    plt.xlabel('Rounds')
    plt.ylabel('L1 Smooth Loss')
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.legend()
    plt.show()
```

```python
plot_ct(SB_filepath, 'Cross Train', 'SB')
```

```python
FLorida_filepath = '/Users/ramanzatsarenko/smores_proj/out/crossval_FL_SB_100epochs_1runs_2025-10-29_12-07-06/metrics.csv'
plot_ct(FLorida_filepath, 'Cross Train', 'FL')
```
