# Collect cross-region rotating metrics
Reads the metrics CSV and shows the selected columns.

Step 2: Last-two holdout


```python
import pandas as pd

CSV_PATHS = [
    "/Users/ramanzatsarenko/smores_proj/out/cross_region_rotating_1runs_2026-01-28_14-45-30/metrics.csv",
    "/Users/ramanzatsarenko/smores_proj/out/cross_region_fl_sb_100epochs_1runs_2026-01-27_18-12-21/metrics.csv",
    "/Users/ramanzatsarenko/smores_proj/out/cross_region_sb_fl_100epochs_1runs_2026-01-27_18-17-09/metrics.csv",
    "/Users/ramanzatsarenko/smores_proj/out/cross_region_100epochs_1runs_2026-01-27_16-54-24/metrics.csv",
    "/Users/ramanzatsarenko/smores_proj/out/cross_region_sb_100epochs_1runs_2026-01-27_17-05-26/metrics.csv"
]
EXP = {
    "/Users/ramanzatsarenko/smores_proj/out/cross_region_rotating_1runs_2026-01-28_14-45-30/metrics.csv" : "FOLDS",
    "/Users/ramanzatsarenko/smores_proj/out/cross_region_fl_sb_100epochs_1runs_2026-01-27_18-12-21/metrics.csv" : "LAST_TWO",
    "/Users/ramanzatsarenko/smores_proj/out/cross_region_sb_fl_100epochs_1runs_2026-01-27_18-17-09/metrics.csv" : "LAST_TWO",
    "/Users/ramanzatsarenko/smores_proj/out/cross_region_100epochs_1runs_2026-01-27_16-54-24/metrics.csv" : "CROSS_REGION",
    "/Users/ramanzatsarenko/smores_proj/out/cross_region_sb_100epochs_1runs_2026-01-27_17-05-26/metrics.csv" : "CROSS_REGION"
}
COLUMNS = ["model", "dataset", "MARE", "MARE_baseline", "SmoothL1Loss()"]

frames = []
for path in CSV_PATHS:
    df = pd.read_csv(path)
    missing = [col for col in COLUMNS if col not in df.columns]
    if missing:
        raise ValueError(f"Missing columns in {path}: {missing}")
    selected = df[COLUMNS].copy()
    selected.insert(0, "source", path)
    selected.insert(2, "experiment", EXP[path])
    frames.append(selected)

res = pd.concat(frames, ignore_index=True)
```

```python
res.to_csv("results_WCCI")
```
